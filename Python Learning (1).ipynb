ELSE STATEMENT

```python
number = 0

while number <20:
    print(number)
    if number == 25:
        break
    number=number+2
else:
    print('no longer <20')
```

CONTINUE

```python
number = 0

while number <20:
    if number == 8:
        continue
    print(number)
    number=number+2
else:
    print('no longer <20')
```

BUILDING A BMI CALCULATOR- PROJECT 1

```python
Name = input("Enter your Name: ")
weight = int(input("Enter your weight in pounds: "))
Height = int(input("Enter your height in inches: "))

BMI = (weight * 703)/(Height * Height)

print(BMI)

if BMI>0:
    if (BMI<18.5):
        print(Name + ", you are Underweight. Kindly check and add your daily calory intake")
    elif (BMI<=24.9):
        print(Name + ", you are Normalweight. This is good. Continue eating healthy and exercising")
    elif (BMI<29.9):
        print(Name + ", you are Overweight. Consider cutting excess fat and calories from your diet")
    elif (BMI<34.9):
        print(Name + ", you are Obese. Kindly seek medical attention from your preferred doctor")
    elif (BMI<39.9):
        print(Name + ", you are SeverelyObese. Make an appointment with your doctor")
    elif (BMI>=40):
        print(Name + ", you are MorbidlyObese. wow")
    else:
        print("Enter valid Input.")
```

BUILDING AN AUTOMATED FILE SORTER IN FILE EXPLORER

```python
import os, shutil
```

```python
path = r"C:/Users/Augustine/Documents/PYTHON TUTS/"
```

* r in this case means read raw

```python
file_name = os.listdir(path)
```

checking if folder exists

```python
Folder_name = ('Png Files','CSV Files','Text Files')
for loop in range(0,3):
    if not os.path.exists(path + Folder_name[loop]):
        #print(path + Folder_name[loop])
        os.makedirs(path + Folder_name[loop])

for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + path):
        shutil.move(path + file,path + "csv files/" + file)
    elif ".png" in file and not os.path.exists(path + "png Files/" + path):
        shutil.move(path + file,path + "png Files/" + file)
    elif ".txt" in file and not os.path.exists(path + "Text Files/" + path):
        shutil.move(path + file,path + "Text Files/" + file)
```

PROJECT; SCRAPING DATA FROM A REAL WEBSITE

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find_all('table',class_='wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
World_Titles = table.find_all('th')

print(World_Titles)
```

```python
World_Table_Titles = [title.text.strip() for title in World_Titles]


print(World_Table_Titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = World_Table_Titles)

df
```

```python
Column_Data = table.find_all('tr')
```

```python
for row in Column_Data[1:]:
    row_data = row.find_all('td')
    Individual_Row_Data = [data.text.strip() for data in row_data]

    
    length = len(df)
    df.loc[length]= Individual_Row_Data
```

```python
df
```
